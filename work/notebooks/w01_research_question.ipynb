## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

I will use the starter dataset to check scale, coverage, and key caveats before modeling.

```python
import pandas as pd

# Starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Real numbers for framing
n_rows = len(df)
n_cols = df.shape[1]
n_clients = df["client_id"].nunique()
avg_pos_zero = (df["avg_position"] == 0).sum() if "avg_position" in df.columns else None
decline_rate = df["is_declining_label"].mean() if "is_declining_label" in df.columns else None

print(f"Rows: {n_rows:,}")
print(f"Columns: {n_cols}")
print(f"Clients: {n_clients}")

if avg_pos_zero is not None:
    print(f"Rows with avg_position == 0 (no data): {avg_pos_zero:,} ({avg_pos_zero/n_rows:.1%})")

if decline_rate is not None:
    print(f"Declining label prevalence: {decline_rate:.1%}")
```

Interpretation (from the numbers above):

- The dataset is large enough for meaningful prioritization work (many pages across multiple clients).
- Multi-client coverage supports a queueing decision use case rather than one-off anecdotal analysis.
- The presence of `avg_position == 0` rows confirms we must treat missing/no-data cases carefully.
- The declining-label prevalence indicates a meaningful minority class to prioritize, which makes ranking useful.
